### 22. What method do we use to join two dataframes.


```python
# Short method
import pandas as pd

df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "id": [1, 2, 4],
    "salary": [50000, 60000, 70000]
})

result = pd.merge(df1, df2, on="id", how="inner")
print(result)
```

```python
# Advance Python
import pandas as pd

def create_dataframe(columns, df_name):
    print(f"\n--- Enter data for {df_name} ---")
    rows = []
    
    while True:
        row = {col: input(f"Enter {col}: ") for col in columns}
        rows.append(row)
        
        cont = input("Add another row? (y/n): ")
        if cont.lower() != 'y':
            break
            
    return pd.DataFrame(rows)


# Define schemas
columns_df1 = ["ID", "Name", "Age"]
columns_df2 = ["ID", "City", "Salary"]

# Create both DataFrames
df1 = create_dataframe(columns_df1, "Personal Info")
df2 = create_dataframe(columns_df2, "Location Info")

# Merge
final_df = pd.merge(df1, df2, on="ID", how="inner")

print("\n--- DataFrame 1 ---")
print(df1)

print("\n--- DataFrame 2 ---")
print(df2)

print("\n--- Final Merged DataFrame ---")
print(final_df)
```

### 31.Write a Python program that checks whether a given year is a leap year or not.

#### Logic (Important for Interviews)
#### A year is a leap year if:

- It is divisible by 4, and

- It is not divisible by 100,
OR

- It is divisible by 400

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

### 33. Write a Python function to find the maximum of three numbers.


```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

result = find_max(10, 25, 15)
print("Maximum number is:", result)
```

### 34. Write a Python function that takes a string as input and returns the reverse of the string.


```python
# using string slicing
def reverse_string(s):
    return s[::-1]

result = reverse_string("Python")
print("Reversed string:", result)
```

```python
# using loop
def reverse_string(s):
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str

print(reverse_string("Python"))
```

```python
#Using Built-in Function
def reverse_string(s):
    return "".join(reversed(s))

s = "Python"
print(reversed(s))
```

### 35. Write a Python function that counts the number of vowels in a given string.


```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    
    for char in s:
        if char in vowels:
            count += 1
    
    return count

# Example usage
result = count_vowels("Data Scientist")
print("Number of vowels:", result)
```

### 38. Write a Python function to count the number of occurence of unique characters in a given String and sort in descending order.

```python
def count_unique_chars(s):
    char_count = {}
    
    # Count occurrences of each character
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    
    # Sort dictionary items by count in descending order
    sorted_chars = dict(sorted(char_count.items(), key=lambda x: x[1], reverse=True))
    
    return sorted_chars

# Example usage
result = count_unique_chars("data science")
print(result)
```

### 39. Write a Python function to remove the duplicate elements from a string without using set concept.

```python
def remove_duplicates(s):
    result = ""
    for char in s:
        if char not in result:
            result += char
    return result

# Example usage
string = "programming"
print("Original String:", string)
print("Without Duplicates:", remove_duplicates(string))
```

### 40. Write a Python function to find the total number of occurence of a substring in a given string.

```python
def count_substring(string, substring):
    return string.count(substring)

# Example usage
text = "data science is fun, data analysis is important"
sub = "data"
print(f"The substring '{sub}' occurs {count_substring(text, sub)} times.")
```
