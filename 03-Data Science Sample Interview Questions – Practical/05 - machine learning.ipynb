## 5. Machine Learning Algorithms


### 41. What is Linear Regression?

**Answer:**  
A method to model the relationship between a dependent variable and one or more independent variables using a straight line.


### 42. What is Logistic Regression used for?

**Answer:**  
For binary classification problems.


### 43. Name three distance-based algorithms.

**Answer:**  
- K-Nearest Neighbors (KNN)  
- K-Means  
- Hierarchical Clustering


### 44. How does KNN work?

**Answer:**  
It classifies data points based on the majority class among their **k nearest neighbors**.


### 45. What is the cost function of Logistic Regression?

**Answer:**  
Log-loss, also known as **Binary Cross-Entropy**.


### 46. What is Regularization?

**Answer:**  
A technique used to penalize large coefficients in regression models to prevent **overfitting**. Common types include **L1 (Lasso)** and **L2 (Ridge)** regularization.


### 47. What is a Decision Tree?

**Answer:**  
A tree-based model that splits data into branches to make predictions or decisions.


### 48. What are Random Forests?

**Answer:**  
An ensemble learning method that combines multiple decision trees to improve accuracy and reduce overfitting for both classification and regression tasks.


### 49. What is Gradient Boosting?

**Answer:**  
A machine learning technique that builds models sequentially, where each new model attempts to correct the errors made by the previous ones.


### 50. What is Cross-Validation?

**Answer:**  
A technique used to evaluate model performance by splitting the dataset into training and validation sets multiple times to ensure the model generalizes well.
