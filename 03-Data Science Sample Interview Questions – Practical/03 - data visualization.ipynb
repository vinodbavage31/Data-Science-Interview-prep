## 24. What is the difference between a bar plot and a histogram?

**A:** Bar plots show categorical data; histograms show frequency of numerical data bins.


## 25. How to identify outliers using box plots?

**A:** Outliers are shown as points outside the whiskers of a box plot.


## 26. How can you handle categorical variables?

**A:** Through one-hot encoding or label encoding.


## 27. How do you deal with duplicates in a dataset?

**A:** Use `df.drop_duplicates()`.


## 28. What does the describe() function do in Pandas?

**A:** Provides summary statistics for numerical columns.


## 29. How to rename a column in Pandas?

**A:** `df.rename(columns={'old': 'new'})`


## 30. What is feature scaling and why is it important?
